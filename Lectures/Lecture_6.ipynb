### Line plots using the plot( ) function

First let's recall how to make a nice plot of the $\sin\theta$ between (0, $2\pi$):

```python
import numpy as np # import the numpy module
import matplotlib.pyplot as plt # import the pyplot sub-module from matplotlib

theta = np.deg2rad( np.arange(0,360,2) )
plt.plot(theta, np.sin(theta), linestyle='-', label='Sine')
plt.show()   # show plot
```

Now, let's make a similar figure plotting the curve of $\cos\theta$ between (0, $2\pi$) using a green dashed curve:

Here I've got two plots, what if I want to show them in the same figure? simply do this:

```python
plt.plot(theta, np.cos(theta), linestyle='--', color='g', label='Cosine')
plt.show() 
```

Then let's put on lables and control the axis to make the plot look much nicer and self-explainable:

```python
import numpy as np # import the numpy module
import matplotlib.pyplot as plt # import the pyplot sub-module from matplotlib

theta = np.deg2rad( np.arange(0,360,8) ) # generate a numpy array for theta

plt.figure(figsize=(12,5))

plt.plot(theta, np.sin(theta), linestyle='-', marker='o',label='Sine') # plot sin(theta)
plt.plot(theta, np.cos(theta), linestyle='--', marker='x',color='g', label='Cosine') # plot cos(theta)
plt.xlim([0,theta.max()]) # set the limits for x axis
plt.ylim([-1.2,1.2])      # set the limits for y axis
plt.xlabel('theta')       # label the x axis
plt.ylabel('Trig Functions')  # label the y axis
plt.title('Sine and Cosine Curves')  # title for the figure
plt.legend()  # show legend

plt.show() # show plot
```

Now let's try plotting $\sin\theta^2$ and $\cos(\theta/2)$ using the same plotting code but specifying different line styles and colors

```python
plt.figure(1,figsize=(16,6))

plt.plot(theta, np.sin(theta**2), color='r', linestyle='-', marker='o', label='Sine', linewidth = 4, markersize = 12)
plt.plot(theta, np.cos(theta/2), 'k-+', label='Cosine', markersize = 10)
plt.xlim([0,theta.max()])
plt.ylim([-1.2,1.2])
plt.xlabel('theta')
plt.ylabel('Trig Functions')
plt.title('Sine and Cosine Curves')
plt.legend()
plt.show() 
```

So here are some useful properties we used in the plt.plot() function to control your plots:

|Property | Values|
|------------ |--------------|
| linewidth   | float value in points |
| linestyle   | [ '-' \| '--' \| '-.' \| ':' \| 'steps' \| ...] |
| marker      | [ '+' \| ',' \| '.' \| 'o' \| 'd' \| '*' \| '1' ] |
| color       | ['b', 'g', 'r', 'k', 'y', 'm', 'c'] or any matplotlib colorm |
| markersize  | float| 
| label       | any string |   

here are some useful functions in plt to control your plots:

More information about controlling line properties can be found here: https://matplotlib.org/tutorials/introductory/pyplot.html#controlling-line-properties

Let's put the two fancy plots within one figure - so you have two panels (subplots) at the same time. To do that, we need to use the subplot function of plt:

```python
plt.figure(figsize=(14, 4)) # start a figure with size 14 (horizontal) and 4 (verticle)

plt.subplot(1, 2, 1) # the first panel 
plt.plot(theta, np.sin(theta), '-', label='Sine', linewidth = 4) # plot sine and cosine
plt.plot(theta, np.cos(theta), 'g--', label='Cosine', linewidth = 0.5)
plt.xlim([0,theta.max()]) # set the limit for the x axis
plt.ylim([-1.2,1.2]) # set the limit for the y axis
plt.xlabel('theta') # label the x-axis
plt.ylabel('Trig Functions') # label the y axis
plt.xticks(np.arange(0,np.pi*2,np.pi/2),('0','90','180','270')) # set the ticks of the x axis
plt.text(3.3,0,'Sine curve') # annotate text at x=3.3, y=0.0
plt.title('Sine and Cosine Curves') # set a title
plt.legend() # show legend
plt.grid() # show grid

plt.subplot(1, 2, 2) # the first panel second
plt.plot(theta, np.sin(theta**2), 'r-o', label='Sine') # make another two plots 
plt.plot(theta, np.cos(theta/2), 'k-+', label='Cosine')
plt.xlim([0,theta.max()])  # set the limit for the x axis
plt.ylim([-1.2,1.2])  # set the limit for the y axis
plt.xlabel('theta') # label the x-axis
plt.ylabel('Trig Functions') # label the y axis
plt.title('Sine and Cosine Curves') # set a title
plt.legend() # show legend

plt.show() # show the plots
```

Now let's try some other 1-D plots using the plt module

### 1. stem( ) and step( ): create a stem plot using you data  

Syntax: **plt.stem(x, y)**, where x is the horizontal axis and y is the vertical axis, for example the following code plots the sine curve using the stem function:

```python
theta = np.deg2rad( np.arange(0,360,8) )

plt.figure(1, figsize = (12,4)) # set figure size

plt.subplot(1,2,1) # the left panel (row 1, colum 1)
plt.stem(theta,np.sin(theta))


plt.subplot(1,2,2) # the right panel (row 1, colum 2)
plt.step(theta,np.sin(theta))

plt.show()
```

More information about the stem() function can be found here: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.stem.html#matplotlib.pyplot.stem

### 2. bar( ): create a bar plot using you 1-D arrays

Syntax: **plt.bar(x, y)**, where x is the horizontal axis and y is the vertical axis, for example the following code plots the sine curve using the stem function:

```python
plt.figure(2, figsize=(14, 4))

theta = np.deg2rad( np.arange(0,360,8) )

plt.subplot(1, 2, 1) # the left panel (row 1, colum 1)
plt.bar(theta,np.sin(theta))

plt.subplot(1, 2, 2) # the right panel (row 1, colum 2)
plt.barh(theta,np.sin(theta))

plt.show()
```

More options on bar plots: 

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

plt.bar(ind, menMeans, width, yerr=menStd, label='Men') # bar plot for the first data set
plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd, label = 'Women')  # bar plot for the second data set

plt.ylabel('Scores') # ylabel
plt.title('Scores by group and gender') # title
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5')) # change xticks to be strings
plt.yticks(np.arange(0, 81, 10)) # fix yticks
plt.legend()

plt.show()
```

### 3. hist( ): create a histgram using you 1-D arrays

Syntax: **plt.hist(x, N)**, where x is the horizontal axis and y is the vertical axis, for example the following code plots the sine curve using the stem function:

```python
plt.figure(3, figsize=(14, 4)) # set up figure size

theta = np.deg2rad( np.arange(0,360,2) )

plt.subplot(1, 2, 1) # left panel
plt.hist(np.sin(theta),5)

plt.subplot(1, 2, 2) # right pabel
plt.hist(np.sin(theta),15)

plt.show()
```

More on the histgrams

```python
x = np.random.randn(10000) # generate 10000 random numbers using the randn() function in the random sub-module of numpy

plt.figure(figsize=(14,4)) # set up figure size

plt.subplot(1,2,1) # left panel
plt.hist(x, bins=50)# histgram with 50 bins

plt.subplot(1,2,2) # right panel
plt.hist(x, bins=10)# histgram with 10 bins

plt.show()
```

### 3. pie( ): create a pie plot using you 1-D arrays

Syntax: **plt.pie(x, N)**. The fractional area of each wedge is given by x/sum(x). If sum(x) < 1, then the values of x give the fractional area directly and the array will not be normalized. The resulting pie will have an empty wedge of size 1 - sum(x).

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels  = ['Na', 'H2', 'O2', 'He','Others']
sizes   = [0.29, 0.22, 0.42, 0.06, 0.01]
explode = (0, 0.1, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.figure(4, figsize=(14, 7)) # generate a figure

plt.subplot(1, 2, 1) # the left panel 
plt.pie(sizes) # generate a simple pie plot

plt.subplot(1, 2, 2) # the right panel
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', # generate a better pie plot with more options
        shadow=True, startangle=0)
plt.title('Atmospheric Concentrations at Mercury') # adding a title

plt.show()
```

### 4. scatter( ): create a scatter plot using you 1-D array(s)

Syntax: **plt.scatter(x, y, s)**. A scatter plot of y vs x with varying marker size and/or color.


```python
planets = ['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
colors = ['b','r','g','k','y','m','c','b']
distance = np.array([0.39,  0.72,  1.  ,  1.52,  5.2 ,  9.58, 19.2 , 30.05]) # normalized to Sun-Earth distance
radius = np.array([0.38,  0.95,  1.  ,  0.53, 11.21,  9.45,  4.01,  3.88]) # normalized to Earth radius
num = np.arange(0,8,1); # index for the planets, from 0 to 7 (total 8)

plt.figure(5,figsize=(12,4)) # setup figure size

plt.subplot(1,2,1) # left panel
plt.scatter(num,distance) # a plain scatter plot

plt.subplot(1,2,2) # right panel
plt.scatter(num,distance,radius*100,color=colors) #scatter plot with scales equal to 100*radius
plt.ylabel('AU')# ylabel
plt.title('Distance From the Sun') # title
plt.xticks(num,planets) # set xticks to be planet names
 
plt.show()
```

### 4. log-scale versus linear scale

Syntax: **plt.semilogx(x, y)**. A scatter plot of y vs x with varying marker size and/or color.



```python
planets = ['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
colors = ['b','r','g','k','y','m','c','b']
distance = np.array([0.39,  0.72,  1.  ,  1.52,  5.2 ,  9.58, 19.2 , 30.05]) # normalized to Sun-Earth distance
radius = np.array([0.38,  0.95,  1.  ,  0.53, 11.21,  9.45,  4.01,  3.88]) # normalized to Earth radius
mass = np.array([0.330,4.87,5.97,0.642,1898,568,86.8,102])

plt.figure(5, figsize=(12,4))

plt.subplot(1,2,1)
plt.scatter(distance,mass,radius*100,color=colors) # generate a scatter plot, 
                                                   # x-axis = distance, y-axis = mass
plt.xlabel('distance (AU)') # xlabel
plt.ylabel('mass (10^6 kg)') # ylabel
plt.title('Distance versus Mass (linear)') # title

plt.subplot(1,2,2)
plt.scatter(distance,mass,radius*100,color=colors)
plt.xlabel('distance (AU)')
plt.ylabel('mass (10^6 kg)')
plt.title('Distance From the Sun (log)')
plt.xscale('log') # switch the x axis to log scale
plt.yscale('log') # switch the y axis to log scale
plt.show()
```
